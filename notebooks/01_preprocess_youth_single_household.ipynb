```python
import glob
import os

all_files = glob.glob(os.path.join(folder_path, "*.xlsx"))

file_list = [
    file for file in all_files
    if "-" in os.path.splitext(os.path.basename(file))[0]
]

print("전체 파일 수:", len(all_files))
print("처리 대상 파일 수:", len(file_list))

for file in file_list:
    print(os.path.basename(file))
```

```python
def preprocess_single_household(file_path, year=2023):

    # 1. 원본을 헤더 없이 읽기
    raw = pd.read_excel(file_path, header=None)

    # 2. '행정기관코드'가 들어 있는 헤더 행 찾기
    header_rows = raw[
        raw.apply(
            lambda row: row.astype(str)
            .str.contains("행정기관코드", na=False)
            .any(),
            axis=1
        )
    ].index

    if len(header_rows) == 0:
        raise ValueError("행정기관코드 헤더를 찾지 못함")

    header_index = header_rows[0]

    # 3. 실제 헤더 기준으로 다시 읽기
    df = pd.read_excel(file_path, header=header_index)

    # 4. 빈 컬럼 제거
    df = df.dropna(axis=1, how="all")

    # 5. 컬럼명 공백·줄바꿈 제거
    df.columns = (
        df.columns.astype(str)
        .str.strip()
        .str.replace("\n", "", regex=False)
        .str.replace(" ", "", regex=False)
    )

    # 6. 필요한 앞 8개 컬럼만 사용
    if df.shape[1] < 8:
        raise ValueError(f"필요한 컬럼 수 부족: {df.shape[1]}개")

    df = df.iloc[:, :8].copy()

    # 구별로 합계 숫자가 컬럼명에 들어가도 강제로 통일
    df.columns = [
        "행정기관코드",
        "행정기관",
        "총거주자수",
        "총연령구간세대수",
        "20~24세",
        "25~29세",
        "30~34세",
        "35~39세"
    ]

    # 7. 행정기관코드 정리
    df["행정기관코드"] = (
        df["행정기관코드"]
        .astype(str)
        .str.replace(".0", "", regex=False)
        .str.strip()
    )

    # 엑셀 아래쪽 빈 행 제거
    df = df[
        df["행정기관코드"].notna()
        & (df["행정기관코드"] != "nan")
    ].copy()

    # 8. 숫자 컬럼 정리
    number_columns = [
        "총거주자수",
        "총연령구간세대수",
        "20~24세",
        "25~29세",
        "30~34세",
        "35~39세"
    ]

    for col in number_columns:
        df[col] = (
            df[col]
            .astype(str)
            .str.replace(",", "", regex=False)
            .str.strip()
        )

        df[col] = pd.to_numeric(df[col], errors="coerce")

    # 9. 자치구 합계행 제거
    # 예: 1126000000
    df = df[
        ~df["행정기관코드"].str.endswith("00000", na=False)
    ].copy()

    # 10. 지역명 분리
    region_split = df["행정기관"].astype(str).str.split()

    df["시도"] = region_split.str[0]
    df["자치구"] = region_split.str[1]
    df["행정동"] = region_split.str[2:].str.join(" ")

    # 11. 연도 추가
    df["연도"] = year

    # 12. 청년 1인세대수 직접 계산
    age_columns = [
        "20~24세",
        "25~29세",
        "30~34세",
        "35~39세"
    ]

    df["청년1인세대수"] = df[age_columns].sum(axis=1, min_count=1)

    # 13. 원본 합계와 직접 계산한 값 비교
    df["합계차이"] = (
        df["총연령구간세대수"]
        - df["청년1인세대수"]
    )

    # 14. 필요한 컬럼 순서 정리
    result_columns = [
        "연도",
        "행정기관코드",
        "시도",
        "자치구",
        "행정동",
        "총거주자수",
        "총연령구간세대수",
        "20~24세",
        "25~29세",
        "30~34세",
        "35~39세",
        "청년1인세대수",
        "합계차이"
    ]

    return df[result_columns]
```

```python
result_list = []
error_list = []

for file_path in file_list:

    try:
        temp_df = preprocess_single_household(
            file_path=file_path,
            year=2023
        )

        result_list.append(temp_df)

        print(
            "처리 완료:",
            os.path.basename(file_path),
            "/ 행 수:",
            len(temp_df)
        )

    except Exception as e:
        error_list.append({
            "파일명": os.path.basename(file_path),
            "오류": str(e)
        })

        print(
            "처리 실패:",
            os.path.basename(file_path),
            "/ 오류:",
            e
        )
```

```python
df_2023 = pd.concat(
    result_list,
    ignore_index=True
)

df_2023.head()
```

```python
def preprocess_year(year):

    folder_path = (
        f"/content/drive/MyDrive/원본 데이터/"
        f"주민등록 1인세대수/{year}년"
    )

    # 폴더 내 모든 엑셀 파일
    all_files = glob.glob(
        os.path.join(folder_path, "*.xlsx")
    )

    # 자치구별 파일은 파일명에 하이픈(-)이 있음
    # 서울 전체 합산 파일은 제외
    file_list = [
        file for file in all_files
        if "-" in os.path.splitext(
            os.path.basename(file)
        )[0]
    ]

    print("=" * 60)
    print(f"{year}년 전체 파일 수:", len(all_files))
    print(f"{year}년 처리 대상 파일 수:", len(file_list))

    result_list = []
    error_list = []

    for file_path in file_list:

        try:
            temp_df = preprocess_single_household(
                file_path=file_path,
                year=year
            )

            result_list.append(temp_df)

            print(
                "처리 완료:",
                os.path.basename(file_path),
                "/ 행 수:",
                len(temp_df)
            )

        except Exception as e:

            error_list.append({
                "연도": year,
                "파일명": os.path.basename(file_path),
                "오류": str(e)
            })

            print(
                "처리 실패:",
                os.path.basename(file_path),
                "/ 오류:",
                e
            )

    if len(result_list) == 0:
        raise ValueError(
            f"{year}년에 처리된 파일이 없습니다."
        )

    year_df = pd.concat(
        result_list,
        ignore_index=True
    )

    error_df = pd.DataFrame(error_list)

    print("-" * 60)
    print(f"{year}년 통합 행 수:", len(year_df))
    print(
        f"{year}년 자치구 수:",
        year_df["자치구"].nunique()
    )
    print(
        f"{year}년 실패 파일 수:",
        len(error_df)
    )

    return year_df, error_df
```

```python
df_2024, error_2024 = preprocess_year(2024)
```

```python
df_2025, error_2025 = preprocess_year(2025)
```
